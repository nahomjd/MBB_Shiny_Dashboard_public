```python
MBB_Opponent_Totals = pd.read_csv('C:\\Users\\jacks\\Documents\\GitHub\\MBB_Shiny_Dashboard\\Source_Data\\MBB_Opponents_Totals.csv')
MBB_Opponent_Totals = MBB_Opponent_Totals.drop(columns=['W', 'G', 'L', 'W/L%', 'W.1', 'Rk'])
MBB_Opponent_Totals
```

```python
MBB_Team_Advanced = pd.read_csv('C:\\Users\\jacks\\Documents\\GitHub\\MBB_Shiny_Dashboard\\Source_Data\\MBB_Season_Advanced.csv')
MBB_Team_Advanced = MBB_Team_Advanced.drop(columns=['W', 'G', 'L', 'W/L%', 'W.1', 'Rk'])
MBB_Team_Advanced
```

```python
MBB_Team_Ratings = pd.read_csv('C:\\Users\\jacks\\Documents\\GitHub\\MBB_Shiny_Dashboard\\Source_Data\\MBB_Ratings.csv')
MBB_Team_Ratings = MBB_Team_Ratings.drop(columns=['W', 'L', 'Rk', 'Unnamed: 3', 'Unnamed: 9', 'Unnamed: 11'])
MBB_Team_Ratings
```

```python
output = db.query(
'''
select * from MBB_Team_totals t
inner join MBB_Opponent_Totals o
on t.Team=o.Team
inner join MBB_Team_Opponent_per_game p
on t.Team=p.Team
inner join MBB_Team_Advanced a
on t.Team=a.Team
inner join MBB_Team_Ratings r
on t.Team=r.School
'''
).to_df()
```

```python
output['NCAA Tournament'] = 0
```

```python
output['NCAA Tournament'] = output['Team'].apply(lambda x: str(x in made_tournament))
```

```python
output
```

```python
db.query(
'''
select Team, School from MBB_Team_totals t
full outer join MBB_Team_Ratings r
on t.Team=r.School
where School is null
or Team is null
'''
).to_df()
```

```python
output['path'] = 'logos\\' + output['Team'].str.replace(' ', '-') + '.png'
output['image_name'] = output['Team'].str.replace(' ', '-').str.lower() + '.png'
```

```python
output.to_csv('C:\\Users\\jacks\\Documents\\GitHub\\MBB_Shiny_Dashboard\\Cleaned_Data\\Season_Stats_23_24.csv')
```
